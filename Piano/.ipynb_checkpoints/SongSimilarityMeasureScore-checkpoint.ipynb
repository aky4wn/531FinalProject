```python
#reading in the data
import pandas as pd
import numpy as np
jupiter = pd.read_csv("CSV/Jupiter.csv", skiprows = 39, header = None).dropna(axis = 0)
jupiter.columns = ["Unknown", "Time", "NoteOnOff", "Channel", "Note", "Velocity"]
```

```python
#empty lists to store pitch and duration

def get1DString(song):
    #need a vector of zeros that is the length of number of note on's...
    n = song.shape[0]
    m = song.groupby("NoteOnOff").count().iloc[1,2]
    pitch = np.zeros(n)
    duration = np.zeros(n)

    #getting all the notes in the song and storing in a dictionary
    possibleNotes = list(set(song["Note"]))
    noteDict = {}
    for i in range(0, len(possibleNotes)):
        noteDict[possibleNotes[i]] = 0

    #looping through the song
    for k in range(0, n):
    
        #if it is the start of a new note
        if song.iloc[k, 2] == ' Note_on_c':
        
            #adding the newest note to the list of pitches
            currentPitch = song.iloc[k,4]
            pitch[k] = (currentPitch)
        
            #storing the beginning location of the note
            noteDict[currentPitch] = k
    
        #if it isn't the start of a new note
        else:
            endNote = song.iloc[k,4]
            noteDuration = k -noteDict[endNote] #right now this just does the number of rows between, how does time play a role?
        
            #finding the location for duration
            for m in range(k, 0, -1):
                if pitch[m] == endNote:
                    duration[m] = noteDuration

    #removing the zeros from the pitch and duration arrays
    pitch = pitch[pitch != 0]
    duration = duration[duration!=0]

    #there are some note off values that come before any note on values that are causing an issue....need to address
    duration = np.append(duration, [0])
    
    #1D string creation
    oneDimString = np.empty(1)
    for a in range(0, len(pitch)):
        lengthVector = np.empty(duration[a])
        lengthVector.fill(pitch[a])
        oneDimString = np.append(oneDimString, lengthVector)
        
    return(oneDimString)
```
